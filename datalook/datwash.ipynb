```python
import glob
import pandas as pd

datlist = glob.glob('../data/*/*/*/*.dat')
df = pd.DataFrame.from_dict({'datlist':datlist}).sort_values('datlist', ignore_index=True)
```

```python
# original path is in column 'datlist_ori', dump path is in column 'datlist'
df['datlist_ori'] = df['datlist']
```

```python
# mending

# mistakenly record falldewideo1_mfy_a1_fall as falldewideo_test_a1_test
df.loc[df['datlist'].apply(lambda x: (x.__contains__('35_falldewideo_fall_test_lights_1000_')) & (x.__contains__('20230612'))), 'datlist'] = df.loc[df['datlist'].apply(lambda x: (x.__contains__('35_falldewideo_fall_test_lights_1000_')) & (x.__contains__('20230612'))), 'datlist'].apply(lambda x: x.replace('35_falldewideo_fall_test_lights_1000_', '35_falldewideo1_fall_mfy_lights_1000_'))

# mistakenly record falldewideo1 as falldewideo
df.loc[df['datlist'].apply(lambda x: (x.__contains__('20230612'))), 'datlist'] = df.loc[df['datlist'].apply(lambda x: (x.__contains__('20230612'))), 'datlist'].apply(lambda x: x.replace('falldewideo_', 'falldewideo1_'))

# keys unmatched
df['datlist'] = df['datlist'].apply(lambda x: x.replace('falldewideo_1', 'falldewideo1'))

df['db'] = df['datlist'].apply(lambda x: x.split('/')[-3].split('_')[1])
df['move'] = df['datlist'].apply(lambda x: x.split('/')[-3].split('_')[2])
df['obj'] = df['datlist'].apply(lambda x: x.split('/')[-3].split('_')[3])
df['light'] = df['datlist'].apply(lambda x: x.split('/')[-3].split('_')[4])
df['angle'] = df['datlist'].apply(lambda x: x.split('/')[-3].split('_')[-2])
df['time'] = df['datlist'].apply(lambda x: x.split('/')[-1].split('.')[0][3:])
df['site'] = df['datlist'].apply(lambda x: x.split('/')[2].split('_')[1])

df['dir'] = df['datlist'].apply(lambda x: x.split('/')[2])

df['key'] = df['dir'] + '_' + df['db'] + '_' + df['move'] + '_' + df['obj'] + '_' + df['light'] + '_' + df['angle'] + '_' + df['site']
```

```python
import copy
df_copy = copy.deepcopy(df)
```

```python
import os
import tqdm

keylist = list(df['key'].value_counts().keys())
for i in tqdm.trange(len(keylist)):
    focus = df.loc[df['key'] == keylist[i]]
    if len(focus) > 5:
        # print(i)
        deletefocus = focus.sort_values('time')[:-5]['datlist']
        # print(len(deletefocus))
        for j in range(len(deletefocus)):
            # os.system('rm %s' % deletefocus.iloc[j])
            # print('%s' % deletefocus.iloc[int(j)])
            df.drop(index=deletefocus.index[j], inplace=True)
    # else:
    #     print(keylist[i])
```

```python
objlist = list(df['obj'].value_counts().index.sort_values())
movelist = list(df['move'].value_counts().index.sort_values())
anglelist = list(df['angle'].value_counts().index.sort_values())
sitelist = list(df['site'].value_counts().index.sort_values())
dblist = list(df['db'].value_counts().index.sort_values())
```

```python
for obj in tqdm.trange(len(objlist), total=len(objlist)):
    for move in range(len(movelist)):
        for angle in range(len(anglelist)):
            for site in range(len(sitelist)):
                for db in range(len(dblist)):
                    focus = df[(df['obj'] == objlist[obj]) & (df['move'] == movelist[move]) & (df['angle'] == anglelist[angle]) & (df['site'] == sitelist[site]) & (df['db'] == dblist[db])].sort_values('time')
                    for i in range(len(focus)):
                        os.system('cp %s %s' % (focus.iloc[i]['datlist_ori'], '../data/csi/%s' % '_'.join(['env' + str(db), 'subj' + str(obj), 'group' + str(move), 'angle' + str(angle), 'rx' + str(site), 't' + str(i)])))
```
